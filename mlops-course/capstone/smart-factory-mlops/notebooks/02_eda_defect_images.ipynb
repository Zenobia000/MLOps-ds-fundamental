# 02 · 瑕疵影像 EDA（產線視覺檢測）

掃描影像目錄建立 manifest（`image_path / label / split`），檢視類別分布。
真實資料用 **MVTec AD**（取得方式見 `data/README.md`）；尚未下載時
`load_images()` 會回傳空 manifest，本 notebook 仍可跑通。

```python
from src.data.loaders import load_images
from src.data.validation import validate_image_manifest

# 預期目錄佈局：<root>/<split>/<label>/*.png（MVTec 風格）。
manifest = load_images()
validate_image_manifest(manifest)
print("manifest 列數：", len(manifest))
manifest.head()
```

```python
# 類別 / split 分布；空 manifest 時提示去 data/README 取得 MVTec AD。
if manifest.empty:
    print("尚無影像資料，請依 data/README.md 下載 MVTec AD 到 data/external/mvtec")
else:
    display(manifest.groupby(["split", "label"]).size().rename("count"))
```
