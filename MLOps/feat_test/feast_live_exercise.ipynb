## 1. prepare dataset and store in parquet format

```python
import pandas as pd
```

```python
data = pd.read_csv('diabetes.csv')
```

```python
data.head()
```

```python
predictors_df = data.loc[:, data.columns != "Outcome"]
target_df = data["Outcome"]

target_df
```

```python
predictors_df.head()
```

```python
timestamps = pd.date_range( end = pd.Timestamp.now(), periods=len(data), freq= 'D').to_frame(name = 'event_timestamp', index = False)

timestamps

```

```python
predictors_df = pd.concat([predictors_df, timestamps], axis=1)

target_df = pd.concat([target_df, timestamps], axis=1)
```

```python
predictors_df.head()
```

```python
target_df.head()
```

```python
dataLen = len(data)

idsList = list(range(dataLen))

idsList
```

```python
patient_ids = pd.DataFrame(data = idsList, columns = ['patient_id'])

patient_ids
```

```python
predictors_df = pd.concat([predictors_df, patient_ids], axis=1)

target_df = pd.concat([target_df, patient_ids], axis=1)
```

```python
predictors_df.head()
```

```python
predictors_df.to_parquet(path="predictors_df.parquet")

target_df.to_parquet(path="target_df.parquet")
```
